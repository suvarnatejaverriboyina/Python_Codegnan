### Tuples

```python
t=(1,2,3,[10,20,30],["hi","hello","namaste","vanakkam"],["raji","roopa"],23)
print(t)
print(t[4][2]+" "+t[5][1])
```

```python
t=(1,2,3,[10,20,30],23)
print(t[3])

print(t)
t[3].append(200) # After Operation it doesnt change the length of the object, remaining positions of other elements.

t[1]=100  #Immutable
print(t)
```

```python
#empty tuple representation
# Single element representation
t=(1)
t1=(1,) # Comma shuld be represented to treat it as tuple otherwise it will read it as integer.
print(t1)
t2=tuple()
print(type(t),type(t1),type(t2))
```

### Methods

```python
#index()
#count() tuple methods
#tuple operations

t=(1,2,3,4,"hi",3.43,4,3,4,1,2)
t1=("helo","nmaste")
print(t.index("hi"))
#print(t.index("hello"))
print(t.count(4))
print(t[::-3])
print(t+t1)
print(t1*2)
```

### Builtin operations

```python
#Builtin operations
t1=(1,2,3,45,5)
t=[122,42,2,422]
print(sum(t1))
print(max(t1))
print(min(t1))
print(len(t1))
print(sorted(t1))   #sorted returns list
print(sorted(t))
print(tuple(t))   #Converts an iterable into tuple
```
