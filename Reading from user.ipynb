```python
###### Output formatting

name="Teja"
Age=21
print(f'My Name is {name} and i am {Age} years Old')    #F format
print("My name is {} and i am {} years Old".format(name,Age))    #Dot format
print("My name is",name,"and i am",Age,"years Old")      #Using Comma
print("My name is "+name+" and i am "+str(Age)+" years old")   #Using string concatenation
```

```python
# Access Specifiers
# string - %s
# integer - %d
# float - %f

Age=21
name="Teja"
print("My name is %s ad i am %d years old."%(name,Age))
print("My name is %s ad i am %d years old."%(Age,name))
```
