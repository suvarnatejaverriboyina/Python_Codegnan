### Dictionary characteristics

```python
# Dictionary is a ordered collection
details={"Name":"Teja","Course":"Python","Batch":57,"Age":21}
print(details)
```

```python
# Dictionary value access using Keys
# Updating Dictionary values
# Adding new items to Dictionary
details={"Name":"Teja","Course":"Python","Batch":57,"Age":21}
print(details["Course"])
#print(details['Location'])   #Key Error: 'Location'
details["Age"]=18
print(details)      #Updated
#Adding new Items
details['Location']="Hyderabad"
print(details)           #Added a new Item
details['Age']=details['Age']+1
print(details)
```

```python
#If multiple keys are present then the dictionary will return only one key,that will be updated with last value.
details={"Name":"Teja","Course":"Python","Batch":57,"Age":21,"Age":25}
print(details)
print(details["Age"])     
```

```python
# Lists cannot be used as Key in Dictionary because list is Unhashable
#Lists can be used as values in Dictionary
details={"Name":"Teja","Course":"Python","Batch":57,"Age":21,"Age":25}
details={"Name":"Teja","Course":"Python",[3,57,53,23]:57,"Age":21,"Age":25}
print(details)   #Type Error
details["course"]=["java","python","django"]  # List as value.
print(details)
```

### Dictionary Methods
- get(key,default_value:Optional): It returns the value if key Exists otherwise it returns default value.
  

```python
details={"Name":"Teja","Course":"Python","Batch":57,"Age":21,"Age":25}
print(details.get("Name"))
print(details.get("marks"))    #If key doenot exist returns None.
print(details.get("marks","Marks Not Available"))   #It key doenot exist returns a specified default value.
```

- Update(new_dictionary): This method updates the values if new_dictioanry keys are already exists in dictionary, otherwise it adds new items to dictionary
  

```python
details={"Name":"Teja","Course":"Python","Batch":57,"Age":21,"Age":25}
details.update({"Name":"Roopa","height":5.1,"Batch":58,"color":"blue"})
details.update({"Age":24})
print(details)
```

- keys(): This method returns the list of Dictionary Keys.
- values(): This method returns the list of Dictionary Values.
- items(): This method returns the list of tuple Dictionary Key-value pairs.

```python
details={"Name":"Teja","Course":"Python","Batch":57,"Age":21,"Age":25}
print(details.keys())
print(details.values())
print(details.items())
```

- popitem(): It returns and removes the last item as tuple.
- pop(key): this removes the specified key item and returns the specified key value, otherwise Keyerror
  

```python
details={"Name":"Teja","Course":"Python","Batch":57,"Age":21,"Age":25}
print(details.popitem())
print(details)
print(details.pop("Course"))
print(details)
print(details.pop("Age"))
```

```python
# Membership Operator : Checks with the key if the key is present in the dictionary returns true elese False.
details={"Name":"Teja","Course":"Python","Batch":57,"Age":21,"Age":25}
print("Raji" not in details)
print("Teja" in details)
print("Name" in details)
print(len(details))
```
