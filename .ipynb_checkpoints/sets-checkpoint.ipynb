### Set Methods

```python
# Element adding method
# add(ele): It adds the element to set if element not in set, if element is present nothing will happen.
s={1,2,3,4,5,4,3}
print(s)
s.add(100)
s.add(5)
s.add('code')
print(s)
```

```python
# Element removing methods
# remove(item): It removes the item if it exists in set, otherwise "Key Error". Why KeyError? Becoz the set elements can be used as dictionary keys.
# discard(item): It removes the item if it exists in set,otherwise nothing.
s={1,2,3,4,5}
s.remove(3)
print(s)
#s.remove(100) if the element is not present in the set remove(item) raises "Key Error"
s.discard(4)
print(s)
#s.discard(100) if the elememt is not present in the set discard(item) does nothing.
```

### Set Operations

```python
#union(set): It returns all the uniques elements from all the sets.
s1={1,3,5,7}
s2={4,5,6,7}
s3={1,2,8,9}
union=s1.union(s2, s3)
print(union)

# Union by using symbol
res=s1 | s2 | s3
print(res)
```

```python
#intersection(set): It returns the common elements from all the sets.
s1={1,3,"hi",5,7,"ji"}
s2={1,4,5,6,"hi",7}
s3={1,2,8,9,"hi"}
intersection=s1.intersection(s2, s3)
print(intersection)
s1.intersection_update(s2, s3)
print(s1)
#By using symbol
#s1={1,3,"hi",5,7,"ji"}
#s2={1,4,5,6,"hi",7}
#s3={1,2,8,9,"hi"}
#res=s1 & s2 & s3
#print(res)
```

```python
# difference(set): It returns the elements of a set those are not in another set.
s1={1,3,"hi",5,7,"ji"}
s2={1,4,5,6,"hi",7}
s3={1,2,8,9,"hi",3}
difference=s1.difference(s2,s3)
print(difference)
s1.difference_update(s2,s3)
print(s1)
# Using the symbol
#res=s1-s2-s3
#print(res)


# we also have difference_update
```

```python
# symmetric_difference(set): It returns all the non common elements
s1={1,3,5,7}
s2={1,4,5,6,7}
s3={1,2,8,9,3,7}
s1.symmetric_difference_update(s2)
print(difference)
s1.symmetric_difference_update(s3)
print(s1)
#print(result)

#Using symbol
#res=s1^s2^s3
#print(res)


#We also have symmetric_difference_update
```

```python
# issubset(set): 
# issuperset(set) :A)_B  A is Superset of B and B is subset.
# isdisjoint(set)
s1={1,2,3}
s2={332,35,2,1,4,6,3,2}
print(s2.issubset(s1))
print(s1.issubset(s2))
print(s2.issuperset(s1))
print(s1.issuperset(s2))
print(s1.isdisjoint(s2))
```
