### String Convertions
- Case Convertion Methods
- upper() : It converts entire string into upper case.
- lower() : It converts entire string into upper case.
- title() : It converts the first letter of every word into uppercase and remaining lowercase in a given string.
- swapcase() : It converts upper to lower and lower to upper.
- capitalize() : It converts first letter into uppercase remainaing into lower case.

```python
#Special characters remains constant no change in that.
string="He sat on the Mat."
print(string.lower())
print(string.upper())
print(string.capitalize())
print(string.swapcase())
print(string.title())
```

#### Remove white spaces or pattern at ends of String
- strip() : By default it removes both the white spaces at the ends of the string.
- lstrip() : It removes the white spaces at the left side of the string.
- rstrip() : It removes the white spaces at the right side of the string.

```python
s="      Codegnan        "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s="@@@@@@Codegn@@@@@an@@@@@@@"
print(s)
print(s.strip("@"))
print(s.lstrip("@"))
print(s.rstrip("@"))
```

```python
# Split() : It splits the entire string into list of substrings based on specified separator.
# by default it splits at space.
string="The Cat sat on the Mat."
print(string.split())
print(string.split("sat"))
print(string.split('a'))
```

```python
# 'sep.join(iterable).Here the iterable can be non-primitive(list,tuple,set).It joins all iterable elements with Separator.
words=['The', 'Cat', 'sat', 'on', 'the', 'Mat.']
print('-'.join(words))
print("".join(words))
print("    ".join(words))
print("_".join(words))
```

```python
# SubString Index Finding
# index(substring) : If substring exists in string, it returns the first occurence of substring index position otherwise 'ValueError'.
# find(substring) : If substring exists in string, it returns the first occurence of substring index position, otherwise -1.
string="The Cat on sat on the Mat."
print(string.index("on"))
print(string.index("z"))
```

```python
string="The Cat on sat on the Mat."
print(string.find("on"))
print(string.find("is"))
print(string.find(" "))
print(string.find(""))
print(string.find("M",-15))
```

```python
# count() : It returns the total non-overlapping occurences of a substring. If the substring doenot exists it return 0.
s="ABcCbABCaBcc"
print(s.count("B"))
print(s.count("AB"))
print(s.count("D"))
lower_string=s.lower()
print(lower_string)
print(lower_string.count('ab'))
```

```python
# replace(old_string,new_string,no_of_strings_to_replace) : It replaces old sub-string with new sub-string.
string="The Cat Sat on the Mat."
print(string.replace("Cat","Man"))
print(string.replace("a","A"))
print(string.replace("a","A",2))
```

### String Checking Methods
- String Checking methods returns the boolean value as a result.
- isupper() : It returns true, if the entire string is in upper case. 
- islower() : It returns true, if the entire string is in lower case. 
- isalpha() : It returns true, if the string contains only alphabets. 
- isdigit() : It returns true, if the string contains only digits. 
- isalnum() : It returns true, if the string contains either alphabets or digits. 
- startswith(substring) : It returns true if the string starts with the given substring.
- endswith(substring) : It returns true if the string ends with the given substring.

### Alignment Methods

```python
string="Codegnan"
print(string.zfill(10))  #takes only width and inserts 0 on left and keep the string right side.# 
print(string.ljust(10,'0')) #align the string to the left and spaces or characters to the right.
print(string.rjust(10,'0')) #align the string to the right and spaces or characters to the left.
print(string.center(10,'0')) #Places the string in the middle and add spaces on the both sides.
```

### Searching

```python
string="codegnan"
print(string.find('e')) #Finds the start index of substring in the given string .if it doesnt exis return -1.
print(string.find('z'))

print(string.rfind('n')) #Finds the index of substring  from right in the given string .if it doesnt exis return -1.
print(string.index('n')) #Finds the start index of the substring ,if it doent exist return "Value error".
print(string.index('z'))
```

### Removing Spaces or Characters

```python
string="   Codeganan.txt   "
print(string.strip()) # Removes spaces on both sides.
print(string.lstrip())
print(string.rstrip())
print(string.lstrip().removeprefix("Code"))
print(string.rstrip().removesuffix(".txt"))
print("Welcome to {}".format(string))
print(f"Welcome to {string}")
```
