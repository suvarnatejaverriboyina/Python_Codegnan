## Type Conversions

```python
# integer to float,boolean,string,complex
num=10
print(float(num))
print(bool(num))
print(str(num))
print(complex(num))
print(complex(num,2))
print(tuple(num))        # Primitive datatypes can't be converted into non primitive like list,dict,set,tuple. 
```

```python
# Float to Integer,string,boolean,complex
num=2.58643
print(int(num))
print(bool(num))
str_val=str(num)
print(str_val[1:])
print(complex(num))
```

```python
num=436
str_val=str(num)
index=str_val.find('.')
print(str_val[:index])
```

```python
# Boolean to String,float,string,complex
b=True
print(int(b))
print(float(b))
print(str(b)[1:3])
print(complex(b))
```

```python
# String to number-float,integer,boolean,list,tuple,set
s="123455656"
print(int(s))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
print(dict(s))  #Dict is not possible
```

```python
# String to number-float,integer,boolean,list,tuple,set
s="Hello"
#print(int(s))
#print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
#print(dict(s))  #Dict is not possible
```

```python
s="123445.4365"
print(int(float(s)))
print(float(s))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
#LIst to String,tuple,set
lst=[1,2,3,4]
print(str(lst)[:3])
print(tuple(lst))
print(set(lst))
l1=[("a",1),("b",2)]
print(dict(l1))
```

```python
#tuple to string,list,set
t=(1,3,4,5,3)
print(str(t)[:4])
print(list(t))
print(set(t))
t1=(["a",1],["b",2])
t2=({"a","b"},{"c","d"})
print(dict(t1))
print(dict(t2))
```

```python
#Set to string,list,tuple
s={1,3,5,2}
print(str(s))
print(list(s))
print(tuple(s))
t1={("a",1),("b",2)}   # Lists are not allowed in sets as they are unhashable
print(dict(t1))
```

```python
dic={"Name":"Roopa","Age":21}
print(str(dic)[:4])
print(list(dic))
print(set(dic))
print(tuple(dic))
```
