### Lists Characteristics

```python
#List is an ordered collection of different datatypes
lst=[1,2,3,42,"hmm","Hello",4.5,657567.346546]
print(lst)
```

```python
# List alllows indexing and slicing.
print(lst[4])
print(lst[-3])
print(lst[:4])
print(lst[::-2])
print(lst[-1::-2])
print(lst[-1:-6:-2])
print(lst[-1:-3:])
print(lst[-4:10])
print(lst[:5:-1])
```

```python
# Empty List Creation
l1=[]
l2=list() #Builtin function
print(type(l1),type(l2))
```

### List operations

```python
# 1. List Concatenation
# 2. List Repetition
l1=[1,2,3]
l2=[3,4,5,6]
print(l1+l2)
print(l1+l1)
print(l1*3)
print([0]*10)
```

### List Methods
- Item adding Methods
  - append(item): It adds element at the end of the list.
  - extend(Iterator): It adds all the iteraator elements at the end of the list.
  - insert(pos,item): It adds element at a specified position.

```python
lst=[10,20,30,40]
lst.append(100)
print(lst)
lst.extend([1,2,3])
print(lst)
lst.append([200,300])
print(lst)
lst.insert(0,10000)
print(lst)
```

### Item Removing Methods
- remove(item): It removes the first occurence item from the list if it exists in the list otherwiswe "value Error".
- pop(pos): It remove and return the specified position item if position is passed, otherwise it remove and returns the last item from the list.
- clear(): It removes all elements form the list.


```python
lst=[1,2,3,4,5,6,2,3,2]
lst.remove(2)
print(lst)
lst.remove(10) #if it doenot contain value error.
```

```python
lst=[1,2,3,4,5,6,2,2,24,23,12]
pos2=lst.pop() # removes last element from the list if index is not passed.
print(pos2,lst)
pos3=lst.pop(4)
print(pos3,lst)
pos1=lst.pop(11) #If the index is out of range it shows index error. 
print(pos,lst)
```

### Index and Count

```python
#index(item): If item is present in list it returns first occurence index otherwise return Value Error.
#count(item): It returns the count of specified element in the list.
lst=[10,20,30,20,40,50,20]
Index=lst.index(20)
print(Index)
Index2=lst.index(90)
print(Index2)
```

```python
c=lst.count(20)
print(c)
```

### Sorting Items

```python
# Sorting Items
# sort(reverse=False): By default it sorts the original list into Ascending Order. If we want in Descending Order sort(reverse=True).
lst=[2,43,554,663,66723]
sorted(lst)
#lst.sort(reverse=True)
print(lst)
lst.sort()
print(lst)
```
